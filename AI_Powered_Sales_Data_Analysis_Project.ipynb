```python
# Importing required Python libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display all columns properly
pd.set_option('display.max_columns', None)

print("Libraries imported successfully")
```

```python
# Reading the uploaded CSV file

try:
    df = pd.read_csv("ai_powered_sales_dataset.csv")
    display(df.head())
except FileNotFoundError:
    print("Error: 'ai_powered_sales_dataset.csv' not found. Please ensure the file was uploaded in the previous step.")
```

```python
# Checking number of rows and columns

print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

```python
# Display first 10 rows

display(df.head(10))
```

```python
# Checking missing values in each column

print("Missing Values in Dataset:")
print(df.isnull().sum())
```

```python
# Checking duplicate rows

duplicate_count = df.duplicated().sum()

print("Total Duplicate Rows:", duplicate_count)
```

```python
# ============================================================
# Data Cleaning
# ============================================================

# Step 1: Remove duplicate rows
df = df.drop_duplicates()

# Step 2: Fill missing Shipping_Cost with median value
if 'Shipping_Cost' in df.columns:
    df['Shipping_Cost'] = df['Shipping_Cost'].fillna(df['Shipping_Cost'].median())

# Step 3: Fill missing Customer_Rating with average rating
if 'Customer_Rating' in df.columns:
    df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].mean())

# Step 4: Convert Order_Date to datetime format
if 'Order_Date' in df.columns:
    df['Order_Date'] = pd.to_datetime(df['Order_Date'])

    # Step 5: Create new date-related columns
    df['Year'] = df['Order_Date'].dt.year
    df['Month'] = df['Order_Date'].dt.month
    df['Month_Name'] = df['Order_Date'].dt.month_name()
    df['Day_Name'] = df['Order_Date'].dt.day_name()

# Step 6: Create Net Profit column
if 'Profit' in df.columns and 'Shipping_Cost' in df.columns:
    df['Net_Profit'] = df['Profit'] - df['Shipping_Cost']

print("Data cleaning completed successfully")
```

```python
# Checking cleaned dataset

print("Missing Values After Cleaning:")
print(df.isnull().sum())

print("\nDuplicate Rows After Cleaning:")
print(df.duplicated().sum())

print("\nUpdated Dataset Shape:")
print(df.shape)

display(df.head())
```

**Exploratory Data Analysis**

```python
# Statistical summary of numerical columns

display(df.describe())
```

```python
# Key business performance indicators

total_revenue = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_net_profit = df['Net_Profit'].sum()
total_orders = df['Order_ID'].nunique()
total_quantity = df['Quantity'].sum()
average_order_value = df['Sales'].mean()

print("========== BUSINESS KPIs ==========")
print(f"Total Revenue: ₹{round(total_revenue, 2)}")
print(f"Total Profit: ₹{round(total_profit, 2)}")
print(f"Total Net Profit: ₹{round(total_net_profit, 2)}")
print(f"Total Orders: {total_orders}")
print(f"Total Quantity Sold: {total_quantity}")
print(f"Average Order Value: ₹{round(average_order_value, 2)}")
```

```python
# Revenue by product category

category_revenue = df.groupby('Product_Category')['Sales'].sum().sort_values(ascending=False)

display(category_revenue)
```

```python
# Profit by product category

category_profit = df.groupby('Product_Category')['Profit'].sum().sort_values(ascending=False)

display(category_profit)
```

```python
# City-wise revenue analysis

city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)

display(city_sales)
```

```python
# Top 10 customers by total sales

top_customers = df.groupby('Customer_Name')['Sales'].sum().sort_values(ascending=False).head(10)

display(top_customers)
```

```python
# Monthly sales trend

monthly_sales = df.groupby('Month_Name')['Sales'].sum()

# Correct month order
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

monthly_sales = monthly_sales.reindex(month_order)

display(monthly_sales)
```

```python
# Region-wise revenue analysis

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

display(region_sales)
```

```python
# Payment mode analysis

payment_mode_count = df['Payment_Mode'].value_counts()

display(payment_mode_count)
```

```python
# Sales channel wise revenue

channel_sales = df.groupby('Sales_Channel')['Sales'].sum().sort_values(ascending=False)

display(channel_sales)
```

**SQL Part in colab**

```python
# Import SQLite library
import sqlite3

# Create database connection
conn = sqlite3.connect("sales_analysis.db")

# Save dataframe as SQL table
df.to_sql("sales_data", conn, if_exists="replace", index=False)

print("Data successfully stored in SQL database")
```

```python
# SQL Query 1: Product category wise total sales

query = """
SELECT Product_Category,
       ROUND(SUM(Sales), 2) AS Total_Sales
FROM sales_data
GROUP BY Product_Category
ORDER BY Total_Sales DESC;
"""

sql_category_sales = pd.read_sql(query, conn)
display(sql_category_sales)
```

```python
# SQL Query 2: City-wise total sales

query = """
SELECT City,
       ROUND(SUM(Sales), 2) AS Total_Sales
FROM sales_data
GROUP BY City
ORDER BY Total_Sales DESC
LIMIT 10;
"""

sql_city_sales = pd.read_sql(query, conn)
display(sql_city_sales)
```

```python
# SQL Query 3: Payment mode wise order count

query = """
SELECT Payment_Mode,
       COUNT(Order_ID) AS Total_Orders
FROM sales_data
GROUP BY Payment_Mode
ORDER BY Total_Orders DESC;
"""

sql_payment_mode = pd.read_sql(query, conn)
display(sql_payment_mode)
```

```python
# SQL Query 4: Monthly revenue

query = """
SELECT Month_Name,
       ROUND(SUM(Sales), 2) AS Total_Sales
FROM sales_data
GROUP BY Month_Name
ORDER BY Total_Sales DESC;
"""

sql_monthly_sales = pd.read_sql(query, conn)
display(sql_monthly_sales)
```

```python
# SQL Query 5: Top 10 customers by sales

query = """
SELECT Customer_Name,
       ROUND(SUM(Sales), 2) AS Total_Sales
FROM sales_data
GROUP BY Customer_Name
ORDER BY Total_Sales DESC
LIMIT 10;
"""

sql_top_customers = pd.read_sql(query, conn)
display(sql_top_customers)
```
